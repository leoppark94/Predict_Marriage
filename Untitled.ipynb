```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 결측치를 확인하여 시각화 하려고 했는데 조회가 되지 않음.
%matplotlib inline
sns.heatmap(dataset1.isnull(), cbar=False)
```

```python
# 'null' 이라는 값이 문자열인거 같아 확인진행하니, 예상대로 문자열이였음
type(dataset1.iloc[0]['birth'])
```

```python
# 결측치가 birth 테이블에 몰려 있는것을 알 수 있음.
%matplotlib inline
birth_replace = {'$null$' : np.NaN}
dataset1['birth'] = dataset1['birth'].replace(birth_replace)
sns.heatmap(dataset1.isnull(), cbar=False)
```

```python
# 기혼과 미혼을 나누어서 결측치 처리할 예정
married_df = dataset1[dataset1['mrg_flg'] == 1]
single_df = dataset1[dataset1['mrg_flg'] == 2]
# 일부 값이 에러나는 경우 NaT 로 치환
married_df['birth'] = pd.to_datetime(married_df['birth'], errors='coerce')
single_df['birth'] = pd.to_datetime(single_df['birth'], errors='coerce')
# 데이터셋을 더 깔끔하게 뭉치게 하기 위해서 birth 에서 year만 남김
married_df['birth_year'] = married_df['birth'].dt.year
single_df['birth_year'] = single_df['birth'].dt.year
married_df = married_df.drop(columns='birth')
single_df = single_df.drop(columns='birth')
# 나누어서 median (중앙값)으로 결측치 처리
married_df = married_df.fillna({'birth_year':married_df.birth_year.median()})
single_df = single_df.fillna({'birth_year':single_df.birth_year.median()})
# 다시 하나의 데이터프레임으로 합침
dataset_marriage = pd.concat([married_df, single_df])
```

```python
# 결측치가 없는것을 확인할 수 있음
%matplotlib inline
dataset_marriage = dataset_marriage.replace('$null$', np.NaN)
sns.heatmap(dataset_marriage.isnull(), cbar=False)
```

```python
dataset_marriage.isna().sum()
```

```python
# 어떤 값들이 있는지 찾아봄
dataset_marriage[dataset_marriage['birth_year'] == 1900]
```

```python
# 해당 데이터셋의 데이터 량이 많지 않고, 
# 1900의 경우 기본 날짜 선택 데이터일 확률이 높음
# 취미코드도 대부분 17 로 선택한 것으로 보아 재대로 선택하지 않았을 수도 있어 해당 데이터는 제거
dataset_marriage = dataset_marriage[dataset_marriage['birth_year'] != 1900]
dataset_marriage.describe()
```

```python
mrg_mapping = {0:0, 1:1, 2:2, 7:0}
House_mapping1 = {"A":0, "B":1, "H":2, "N":3, "V":4, "X":5, "Z":6}
House_mapping2 = {0:0, 1:1, 2:2, 3:3, 4:4, 5:0, 6:0, 7:0, 8:0}

dataset_marriage['mrg_flg'] = dataset_marriage['mrg_flg'].replace(mrg_mapping)
dataset_marriage['h_type1'] = dataset_marriage['h_type1'].replace(House_mapping1)
dataset_marriage['h_type2'] = dataset_marriage['h_type2'].replace(House_mapping2)
```

```python
all_dataset = pd.concat([dataset_marriage, dataset2], axis=1, join='inner')
```

```python
# 데이터를 min-max 정규화를 통해 정규화 진행
all_dataset = (all_dataset - all_dataset.min())/(all_dataset.max() - all_dataset.min())
all_dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
# x 와 y 분리
train_x, train_y = all_dataset.drop(columns='mrg_flg'), pd.DataFrame(all_dataset.mrg_flg)
# X_train, X_test, Y_train, Y_test = train_test_split(train_x, train_y, test_size=0.2, random_state=20221216)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
import warnings

warnings.filterwarnings(action='ignore')
```

```python
# KNN
clf = KNeighborsClassifier(n_neighbors = 5)
scoring = 'accuracy'
score1 = cross_val_score(clf, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(score1)*100, 2)
```

```python
# Decision Tree
clf = DecisionTreeClassifier()
score2 = cross_val_score(clf, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(score2)*100, 2)
```

```python
# RandomForest
clf = RandomForestClassifier(n_estimators=20)
score3 = cross_val_score(clf, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(score3)*100, 2)
```

```python
# Naive Bayes
clf = GaussianNB()
score4 = cross_val_score(clf, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(score4)*100, 2)
```

```python
clf = XGBClassifier()
score6 = cross_val_score(clf, train_x, train_y, cv=k_fold,scoring=scoring)
round(np.mean(score6)*100, 2)
```

```python
clf = LGBMClassifier()
score7 = cross_val_score(clf, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(score7)*100, 2)
```

### 앙상블 적용

```python
XG = XGBClassifier()
LG = LGBMClassifier()

voting = VotingClassifier(estimators=[('XGBoost',XG), ('LGBM',LG)], voting='soft')

final_result = cross_val_score(voting, train_x, train_y, cv=k_fold, scoring=scoring)
round(np.mean(final_result)*100, 2)
```
